```python
import pandas as pd

df = pd.read_csv('household_power_consumption.txt', sep=';', low_memory=False, na_values='?')

print("Primeiras linhas do DataFrame:")
print(df.head())

print("\nInformações do DataFrame:")
print(df.info())

print("\nQuantidade de valores faltantes antes do tratamento:")
print(df.isnull().sum())


colunas_numericas = df.columns.drop(['Date', 'Time'])


df[colunas_numericas] = df[colunas_numericas].apply(pd.to_numeric, errors='coerce')


df[colunas_numericas] = df[colunas_numericas].fillna(df[colunas_numericas].median())

print("\nQuantidade de valores faltantes após o preenchimento:")
print(df.isnull().sum())


df_subset = df[['Global_active_power', 'Global_intensity', 'Sub_metering_3']]


print("\nMédia das colunas selecionadas:")
print(df_subset.mean())

print("\nMediana das colunas selecionadas:")
print(df_subset.median())

print("\nDesvio padrão das colunas selecionadas:")
print(df_subset.std())

df_amostra = df_subset.sample(frac=0.01, random_state=42)

print("\nMédia da amostra:")
print(df_amostra.mean())

print("\nMediana da amostra:")
print(df_amostra.median())

print("\nDesvio padrão da amostra:")
print(df_amostra.std())
```
